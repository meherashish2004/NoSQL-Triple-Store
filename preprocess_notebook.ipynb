```python
from datetime import datetime
import pandas as pd

df = pd.read_csv("./yago_uniques.csv")
df.head()
```

```python
df.sort_values(by=['subject', 'predicate'], inplace=True)
df.head()
```

```python
df.head(30)
```

```python
print(len(df))
print(len(df)/3)
```

```python
num_partitions = 3

# Calculate the number of rows per partition
rows_per_partition = len(df) // num_partitions

# Partition the DataFrame and export to CSV files
for i in range(num_partitions):
    # Calculate the start and end indices for the current partition
    start_idx = i * rows_per_partition
    end_idx = (i + 1) * rows_per_partition if i < (num_partitions - 1) else len(df)
    
    # Slice the DataFrame to extract rows within the current partition range
    partition_df = df.iloc[start_idx:end_idx]
    
    # Export the partition DataFrame to a CSV file
    partition_df.to_csv(f'partition_{i + 1}.csv', index=False)

    # Optionally, you can display the partition DataFrame
    print(f"DataFrame for partition {i + 1}:")
    print(len(partition_df))
```

```python
del df
del partition_df
```

```python
for i in range(num_partitions):
    if i < num_partitions-1:
        df_partitioni = pd.read_csv(f'./partition_{i + 1}.csv')
        df_partitioniplus1 = pd.read_csv(f'./partition_{i + 2}.csv')
    else:
        df_partitioni = pd.read_csv(f'./partition_{i + 1}.csv')
        df_partitioniplus1 = pd.read_csv(f'./partition_{(i + 2)%3}.csv')


    concatenated_df = pd.concat([df_partitioni, df_partitioniplus1])
    concatenated_df['timestamp'] = datetime.min.strftime("%Y-%m-%d %H:%M:%S.%f")
    concatenated_df.to_csv(f'node_sharded_yagodataset{i + 1}.csv', index=False)

    # Optionally, you can display the partition DataFrame
    print(f"DataFrame for node {i + 1}:")
    print(len(concatenated_df))
```
